```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/house_data.csv')
df.head()
```

```python
df.info()
```

```python
x = df[['sqft_living']]
y = df['price']

x.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X=x , y=y)
```

```python
slope = model.coef_
intercept = model.intercept_

print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")
```

```python
x_range = np.linspace(x.min(), x.max(), 100)
y_range = model.predict(x_range.reshape(-1, 1))  # Corrected line

plt.scatter(x, y, label="Data", color="b", alpha=0.7)

plt.plot(x_range, y_range, label="Regression Line", color="r")

plt.xlabel("Sqft Living")
plt.ylabel("Price")
plt.title("Linear Regression")

plt.legend()
plt.grid(True)
plt.show()
```

```python
def predict_price(sqft_living):
  predicted_price = model.predict([[sqft_living]])
  return predicted_price[0]

sqft_living = 2000
predicted_price = predict_price(sqft_living)
print(f"Predicted price for a house with {sqft_living} sqft living space: ${predicted_price:.2f}")
```
